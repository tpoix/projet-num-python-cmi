```python
from itertools import product
masses = {
    "H": 1,
    "C": 12,
    "O": 16,
    "N": 14
}

molecule = 4 * masses["H"] + masses["C"] + masses["O"]

for nbH in range(11):
    masse_h = nbH * masses["H"]
    for nbC in range(1,3):
        masse_c = nbC * masses["C"]
        for nbO in range(3):
            masse_O = nbO * masses["O"]
            for nbN in range(3):
                masse_N = nbN * masses["N"]
                masse_totale = masse_h + masse_c + masse_O + masse_N
                if(masse_totale == molecule):
                    print(f"Molécule: H {nbH} C {nbC} O {nbO} N {nbN}")

ranges = [range(11), range(1,3), range(3), range(3)]
for nuplet in product(*ranges):
    masse_totale = sum([nuplet[i] * list(masses.values())[i] for i in range(len(nuplet))])
    if(masse_totale == molecule):
        print(f"Molécule: {nuplet}")
```
